```python
from matplotlib import pyplot as plt
```

```python
# plt.style.available
```

```python
# Developer salaries by age
dev_x = []
for i in range(11):
    dev_x.append(i + 25)
dev_y = [38496, 42000, 46752, 49320, 53200, 56000,
         62316, 64928, 67317, 68748, 73752]

# Python developer salaries by age
# py_dev_x = []
# for i in range(11):
#     py_dev_x.append(i + 25)
py_dev_y = [45372, 48876, 53850, 57287, 63016, 65998, 70003,
            70000, 71496, 75370, 83640]

# Javascript developer salaries
js_dev_y = [37810, 43515, 46823, 49293, 53437, 56373,
            62375, 66674, 68745, 68746, 74583]
```

```python
def visualize(style):
    # plt.xkcd()
    plt.style.use(style)
    plt.plot(dev_x, py_dev_y, color = '#bb00ff',linewidth = '3', label = 'python')
    plt.plot(dev_x, js_dev_y, color = '#00aaff',linewidth = '3',  label = 'javascript')
    plt.plot(dev_x, dev_y, color = 'k', linestyle = '--', marker = '.',
            label = 'all devs')

    plt.xlabel('Ages')
    plt.ylabel('Salary')
    plt.title(style)

    # plt.legend(['all devs', 'python'])
    plt.legend()

    plt.grid(True)
    plt.tight_layout()
    plt.savefig('plot1.png')
    plt.show()

visualize('fivethirtyeight')
```
