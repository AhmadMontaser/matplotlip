```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('INC 5000 Companies 2019.csv')
```

```python
data.head()
```

```python
foundation = data['yrs_on_list']
previous_workers = data['previous_workers']
workers = data['workers']

plt.plot(foundation, previous_workers, color = '#bd0dd1',
         linestyle='--', label = 'previous_workers')

plt.plot(foundation, workers, color = '#12c9de',
         linestyle='--', label = 'workers') 

plt.legend()
plt.title('Companies')
plt.show()
```

```python
# Developer salaries by age
dev_x = []
for i in range(11):
    dev_x.append(i + 25)
dev_y = [38496, 42000, 46752, 49320, 53200, 56000,
         62316, 64928, 67317, 68748, 73752]

# Python developer salaries by age
# py_dev_x = []
# for i in range(11):
#     py_dev_x.append(i + 25)
py_dev_y = [45372, 48876, 43850, 47287, 63016, 65998, 70003,
            70000, 71496, 75370, 83640]

# Javascript developer salaries
js_dev_y = [40000, 43515, 46823, 49293, 43437, 56373,
            62375, 66674, 68745, 68746, 74583]
```

```python
plt.plot(dev_x, py_dev_y, color = '#bb00ff',linewidth = '3', label = 'python')
plt.plot(dev_x, js_dev_y, color = '#ff23e0',linewidth = '3',  label = 'javascript')
# plt.plot(dev_x, dev_y, color = 'k', linestyle = '--',
        # label = 'all devs')

overall_med = 67000

dev_x2 = pd.Series(dev_x)
py_dev_y2 = pd.Series(py_dev_y)
js_dev_y2 = pd.Series(js_dev_y)

plt.title('title')

plt.fill_between(dev_x2, py_dev_y2, js_dev_y2,
                  where=(py_dev_y2 >= js_dev_y2),
                   interpolate=True, color='skyblue', alpha=0.25, label='py_ab')

plt.fill_between(dev_x2, py_dev_y2, overall_med,
                  where=(py_dev_y2 <= js_dev_y2),
                   interpolate=True, color='red', alpha=0.25, label='js_ab')

plt.legend()

plt.tight_layout()
plt.savefig('plot1.png')
plt.show()
```
