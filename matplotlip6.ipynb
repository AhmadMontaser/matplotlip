```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
```

```python
plt.hist(ages, bins=5, edgecolor='black')
plt.title('title')
plt.xlabel("Age")
plt.ylabel("respondents")
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('INC5000Companies2019.csv')
data.head()
```

```python
foundation_year = data['founded']
# workers_number = data['workers']
```

```python
bins = [1990, 1995, 2000, 2005, 2010, 2015, 2020]
plt.hist(foundation_year, bins=bins, edgecolor='black')
plt.title('title')
plt.xlabel("foundation_year")
plt.ylabel("number")
plt.tight_layout()
plt.show()
```

```python
bins = [1990, 1995, 2000, 2005, 2010, 2015, 2020]
plt.hist(foundation_year, bins=bins, edgecolor='black', log=True)
plt.title('title')
plt.xlabel("foundation_year")
plt.ylabel("number")
plt.tight_layout()
plt.show()
```

```python
median_year = 2009

bins = [1990, 1995, 2000, 2005, 2010, 2015, 2020]

plt.hist(foundation_year, bins=bins, edgecolor='black')

plt.axvline(median_year, color='#aa00ff', label='Median', linewidth=2)

plt.legend()
plt.title('title')
plt.xlabel("foundation_year")
plt.ylabel("number")
plt.tight_layout()
plt.show()
```
