```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47466]
labels = ['python', 'js', 'c#']
colors = ['#008fd5', '#fc4f30', '#e5ae37']

plt.pie(slices, labels=labels,colors=colors, wedgeprops={'edgecolor': 'black'})

plt.title("piechart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47466]
labels = ['python', 'js', 'c#']
colors = ['#008fd5', '#fc4f30', '#e5ae37']
explode = [0, 0.05, 0]

plt.pie(slices, labels=labels,colors=colors, explode=explode,
        shadow=True, startangle=200,
        autopct='%1.1f%%',
         wedgeprops={'edgecolor': 'black'})

plt.title("piechart")
plt.tight_layout()
plt.show()
```
